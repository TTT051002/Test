```python
import os
import pandas as pd
import numpy
path = 'Data'

#Tạo một file rỗng
All_Thang = pd.DataFrame()

#đọc dữ liệu từng file và tổng hợp thành 1 file
for file in os.listdir(path):
    path_file = path + '/' + file
    current_data = pd.read_csv(path_file)
    print(file, '---', current_data.shape[0])
    
    All_Thang = All_Thang.append(current_data)
```

# 2 LÀM SẠCH DỮ LIỆU 

* 2.1 Xử lý các dòng không có giá trị 

```python
#Tạo một bản sao của tập dữ liệu thô ban đầu:
all_data = All_Thang.copy()
all_data.info()
```

```python
All_Thang.shape
```

```python
#kiểm tra số lượng bản ghi null trong dữ liệu:
All_Thang.isnull().sum()
```

```python
#Liệt kê 1 số bản ghi NAN
nan_df = all_data[all_data.isna().any(axis=1)]
display(nan_df.head())
print('Tổng số dòng không có dữ liệu:', nan_df.shape[0])
```

```python
#Xóa tất cả các dòng dữ liệu rỗng
all_data.dropna(how='all',inplace=True)

#Reset lại chỉ số cột
all_data.reset_index(drop=True,inplace=True)

all_data.info()
```

* 2.2 Xử lý các dòng tiêu đề trong dữ liệu 

```python
all_data.iloc[9310:9320,:]
```

```python
#Liệt kê các dòng tiêu đề trong Data:
all_data[all_data['Order ID']=='Order ID']
```

```python
#Lọc và loại bỏ tất cả các dòng tiêu đề trong Data:
all_data = all_data[all_data['Order ID'] != 'Order ID']

#Reset lại index của DataFrame:
all_data.reset_index(drop=True,inplace=True)

all_data.info()
```

* 2.3 Chuyển đổi các dữ liệu cột số 

```python
#Chuyển dổi kiễu dữ liệu sang dạng số:
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])

#Chuyển đổi kiểu dữ liệu sang dạng time:
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])

all_data.info()
```

```python
#Hiển thị 5 dòng đầu
all_data.head()
```

* 2.4 Thêm các cột dữ liệu mới phục vụ phân tích

```python
all_data['Day'] = all_data['Order Date'].dt.month
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data.head()
```

```python
all_data.tail()
```

## Tách tên cột thành phố vào cột mới 

```python
def get_city(address):
    return address.split(",")[1].strip(" ")

def get_state(address):
    return address.split(",")[2].split(" ")[1]

all_data['city'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
all_data.head()
```

## Cột Sales : Được tính bằng số lượng sản phẩm * Giá bán sản phẩm 

```python
all_data['Sales'] = all_data['Quantity Ordered'].astype('int') * all_data['Price Each'].astype('float')
all_data.head()
```
